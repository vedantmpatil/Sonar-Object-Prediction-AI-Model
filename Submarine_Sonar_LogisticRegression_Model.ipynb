**Lets Create a Model for a Submarine using Sonar Technology to Identify or Distinguish between a Rock and Mine from the Sonar Inputs.**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonarDf = pd.read_csv('/content/sonar data.csv',header = None)
sonarDf
```

```python
sonarDf.isnull().sum()
```

```python
sonarDf.shape
```

```python
sonarDf[60].value_counts()
```

```python
X = sonarDf.drop(columns=60,axis=1)
Y = sonarDf[60]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
LRModel = LogisticRegression()
```

```python
LRModel.fit(X_train,Y_train)
```

```python
X_train_predicted = LRModel.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predicted,Y_train)
```

```python
print(training_data_accuracy)
```

```python
X_test_predicted = LRModel.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predicted,Y_test)
```

```python
print(test_data_accuracy)
```
